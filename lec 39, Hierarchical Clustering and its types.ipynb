# 3. What is Hierarchical Clustering?
Hierarchical clustering is an unsupervised machine learning algorithm used to group similar data points into clusters based on a hierarchy. It builds a tree-like structure (called a dendrogram) to represent nested clusters at different levels of similarity.


# Two Types of Hierarchical Clustering

# 1. Agglomerative (Bottom-Up)

  a) Starts with each point as its own cluster.

  b) Iteratively merges the closest clusters until all points belong to a single cluster.




# 2. Divisive (Top-Down)

  a) Starts with all points in one cluster.

  b) Recursively splits the cluster into smaller clusters based on dissimilarity.
  
  
   
  

# Steps in Agglomerative Hierarchical Clustering

1. Calculate distance between every pair of points (Euclidean, Manhattan, etc.).

2. Assign each point to its own cluster.

3. Merge the closest clusters (based on linkage criteria like single, complete, average).

4. Repeat until all data points are merged into one big cluster (or desired number of clusters).

5. Cut the dendrogram at a suitable level to choose the number of clusters.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate sample data
X, _ = make_blobs(n_samples=100, centers=3, random_state=42)


plt.scatter(X[:, 0], X[:, 1], cmap='rainbow')
plt.title("Agglomerative Hierarchical Clustering")
plt.show()
```

```python
# Fit hierarchical clustering
model = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels = model.fit_predict(X)
labels
```

```python
# Plot clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
plt.title("Agglomerative Hierarchical Clustering")
plt.show()
```

```python
# Plot dendrogram
linked = linkage(X, 'ward')
plt.figure(figsize=(10, 5))
dendrogram(linked)
plt.title("Dendrogram")
plt.xlabel("Data points")
plt.ylabel("Distance")
plt.show()
```
